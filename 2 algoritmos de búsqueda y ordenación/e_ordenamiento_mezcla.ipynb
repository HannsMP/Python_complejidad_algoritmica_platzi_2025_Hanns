```python
from random import randint
```

```python
def ordenamiento_mezcla(lista:list):
    n = len(lista)

    if n > 1:
        medio = n // 2

        # subdivide las lista
        izquierda, derecha = lista[:medio], lista[medio:]
        print('->', izquierda, ',' ,derecha)
        
        len_izquierda, len_derecha = len(izquierda), len(derecha)

        ordenamiento_mezcla(izquierda)
        ordenamiento_mezcla(derecha)

        i, j, k = 0, 0, 0

        # comparacion
        while i < len_izquierda and j < len_derecha:
            val_izquierda = izquierda[i]
            val_derecha = derecha[j]
            if val_izquierda < val_derecha:
                lista[k] = val_izquierda
                i += 1
            else:
                lista[k] = val_derecha
                j += 1
            
            k += 1

        print(lista)

        while i < len_izquierda:
            lista[k] = izquierda[i]
            i += 1
            k += 1

        while j < len_derecha:
            lista[k] = derecha[j]
            j += 1
            k += 1
            
        print(f'  izquierda {izquierda}, derecha {derecha}')
        print(lista)
        print('-' * 50)

    return lista
```
